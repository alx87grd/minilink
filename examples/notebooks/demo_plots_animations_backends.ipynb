```python
import numpy as np

from minilink.dynamics.catalog.pendulum.pendulum import Pendulum
```

# Animation tools

```python
# Plant system
sys = Pendulum()
sys.params["m"] = 1.0
sys.params["l"] = 1.0
sys.x0[0] = 2.0
```

```python
# Natural response
# sys.compute_trajectory(tf=20.0, dt=0.01, show=False)
```

```python
# Forced response
traj = sys.compute_forced(u=lambda t: 10.0 * np.sin(10.0 * t), tf=3.0)
```

```python
fig = sys.plot_trajectory()
```

```python
fig = sys.plot_trajectory(signals=("x", "u"), backend="plotly")
```
